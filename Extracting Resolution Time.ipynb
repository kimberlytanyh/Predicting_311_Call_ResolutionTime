```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
dataset = pd.read_csv('data_no_na.csv',sep=',',encoding="utf-8")
```

```python
dataset['opened']=pd.to_datetime(dataset['opened'])
dataset['closed']=pd.to_datetime(dataset['closed'])
dataset['difference'] = [(dataset['opened'][i] - dataset['closed'][i]) for i in range(len(dataset['opened']))]
```

```python
dataset['difference_mins'] = [int(dataset['difference'][i].total_seconds()/60) for i in range(len(dataset['difference']))]
```

```python
dataset['difference_mins']
```

```python
dataset['difference_mins'] = abs(dataset['difference_mins'])
```

```python
dataset['difference_mins']
```

```python
dataset.drop('time_mins',axis=1)
```
