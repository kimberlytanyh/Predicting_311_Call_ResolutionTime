```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("dataset_time_difference.csv")
```

```python
#removing case ID

df = df.drop(columns= ["caseID"])
```

```python
#extract the month from open columns

df["opened"] = df['opened'].map(lambda date: pd.Timestamp(date))
df["month"] = df['opened'].map(lambda dm: dm.month)
df
```

```python
df.month.value_counts()
```

```python
#delete the open and closed columns

df = df.drop(columns= ["opened","closed"])
```

```python
#Reduce the data to top 5 Categories
new_df = df[df.category.isin(['Street and Sidewalk Cleaning','Parking Enforcement','Graffiti','General Request - PUBLIC WORKS','Illegal Postings'])]
```

```python
new_df
```

```python
# performing mean target encoding on neighborhood to prevent curse of dimensionality
# including smoothing (to reduce overfitting and data leakage)

target_mean = new_df["difference_mins"].mean()
agg = new_df.groupby("neighborhood")["difference_mins"].agg(["count", "mean"])
counts = agg["count"]
means = agg["mean"]
weight = 100

# compute smoothed means
smooth = (counts * means + weight * target_mean)/ (counts + weight)

# replace neighborhood values with smoothed mean
new_df["neighborhood"] = new_df["neighborhood"].map(smooth)
```
